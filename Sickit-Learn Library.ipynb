# Now we are Ready :D

### 1) Construct the features dataFrame and target dataFrame

```python
features = housing_data.drop('median_house_value', axis=1)
features
```

```python
target = housing_data['median_house_value']
target
```

### 2) Split your dataset into training dataset and testing dataset, its most common to use 80% for training and 20% for testing

```python
from sklearn.model_selection import train_test_split
features_train, features_test, target_train, targe_test = train_test_split(features, target, test_size=0.2)
```

```python
features_train.shape, features_test.shape
```

### 3) use sickit-learn APIs to build your model 

```python
# => Step [1] : Import the Linear regression model/Estimator from sklearn.linear_model
from sklearn.linear_model import LinearRegression

# => Step [2] : instantiate a linear Regression Object
## QUESTION (1) 
#### I know you may wonder what is the meaning of (normalize = True) ???
#### ANSWER : normalize scales all numeric features to be between 0 and 1, having all features in the same scale can improve the performance of your ML model
## QUESTION (2)
#### so now tell me what is the usage of .fit() ?
#### ANSWER : the .fit() is API to train your model with provided features_Train and target_Train datasets
linear_model = LinearRegression(normalize = True).fit(features_train,target_train)
```

#### Congrats, you trained your first ML model ^_^ 

### 4) Lets now Evaluate your ML model

```python
Training_DataSet_Score = linear_model.score(features_train,
                                            target_train)
print("Training Score on the Training dataset : ", Training_DataSet_Score)
```

```python
# i will assign all used features to build our model
predictors = features_train.columns
#predictors.size # = 13
predictors
```

##### Remember the linear regression EQN ?
###### `y = wx + b` where `w` is the coeff of each feature which represents the effect of each feture to our predicted result, if `w is +` so if this feature increase, the predicted value will be closer to the actual value, the opposite when `w is -`

```python
# Lets explore the coeff that our model used 
coeff = pd.DataFrame(linear_model.coef_, predictors, columns=['Coeff']).sort_values(by='Coeff', ascending=True)
coeff
```

### 5) start use your fully trained model to predict something with a new dataset

```python
# use mode.predict(NewDataSet)
predicted_House_Price = linear_model.predict(features_test)
# lets make a dataframe to easily compare between the actual and predicted values
predicted_vs_actual = pd.DataFrame({
    "Predicted" : predicted_House_Price,
    "Actual" : targe_test
})
# explore random 10 rows of this dataframe to see the results
predicted_vs_actual.sample(10)
```

### 6) Evaluate your model

#### usually R2 is a good evaluator, R2 also called least square error

```python
from sklearn.metrics import r2_score
R2_metric = r2_score(targe_test, predicted_House_Price)
R2_metric
```

#### From the Above value, you can observe that your model behave with new data < its behave with trained data.. this is not good btw

### Lets plot actual vs predicted
##### you know that when we use plt.plot() the default graph will be a line plot, right !? study matplotlib and you will know that.

```python
fig, ax = plt.subplots(figsize = (12, 10))
plt.plot(predicted_vs_actual['Predicted'], label = "Predicted")
plt.plot(predicted_vs_actual['Actual'], label = "Actual")
plt.legend()
plt.show()
```
